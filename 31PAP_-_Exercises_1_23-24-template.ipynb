# Truth Table

You are hired to build a program that returns the truth of a logical implication statement. An implication or conditional statement is a declarative sentence composed of two statements $p$ and $q$ connected by an *if .. then* phrase. It is something of the form "if $p$ then $q$". For example, you can say "if it rains then I will get wet", where $p =$ "it rains" and $q = $ "I will get wet".

An logical implication statement is written like $p \rightarrow q$. We can verify the truth value of a statement by looking at the implication truth table:

| **p** | **q** | **p &rarr; q** |
|:-----:|:-----:|:--------------:|
|True |True |True |
|True |False|False|
|False|True |True |
|False|False|True |
